## Advance operation with Real life Example 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import numpy as np

# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("====  Zomanto Sales Analysis ==== ")
print("\n sales data shape :", sales_data.shape)
print("\n Sample data for 1st 3 restaurants : \n", sales_data[0:3])
```

```python
# Total sales for each year
print(np.sum(sales_data, axis=0))
yearly_sales = np.sum(sales_data[:1:], axis=0)
print("\n Total sales for each year : \n", yearly_sales)
print(yearly_sales)


# MINMUM sales for each year
min_sales = np.min(sales_data[: , 1 :] , axis=1)
print( min_sales)


# MAXMUM sales for each year
max_sales = np.max(sales_data[: , 1 :] , axis=0)
print( max_sales)


# AVERAGE sales for each year
avg_sales = np.mean(sales_data[: , 1 :] , axis= 1)
print( avg_sales)
```

```python
cumulative_sales = np.cumsum(sales_data, axis=1)
print(cumulative_sales)

plt.figure(figsize=(5, 4))
plt.plot(cumulative_sales[:, 0], label='2021')
plt.plot(cumulative_sales[:, 1], label='2022')
plt.plot(cumulative_sales[:, 2], label='2023')
plt.plot(cumulative_sales[:, 3], label='2024')
plt.xlabel('Restaurant ID')
plt.ylabel('Cumulative Sales')
plt.title('Cumulative Sales Over Time')
plt.legend()
plt.show()
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([5, 4, 3, 2, 1])

print("vector addition :",vector1 + vector2)   
print("vector substraction :",vector1 - vector2)
print("vector multiplication :",vector1 * vector2) 
print("vector dot product :",np.dot(vector1,vector2))

angle = np.arccos(np.dot(vector1,vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("angle between vector1 and vector2 : ", angle)
```
